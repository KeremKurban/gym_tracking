```python
from functools import reduce
import numpy as np
import scipy
from numpy.random import randn
import matplotlib.pyplot as plt
```

```python
def multi_conv(*arrs, conv = lambda x, y: np.convolve(x, y, mode='full')):
    return reduce(conv, arrs)
```

```python
n=100
x=np.random.normal(n,size=100)
y=randn(n)+0.2
z=randn(n)*0.3
```

```python
plt.plot(sorted(x))
```

```python
mu1, sigma1 = 420, 40 # mean and standard deviation
mu2, sigma2 = 530, 35 # mean and standard deviation
mu3, sigma3 = 560, 35 # mean and standard deviation

s1 = np.random.normal(mu1, sigma1, 1000)
s2 = np.random.normal(mu2, sigma2, 1000)
s3 = np.random.normal(mu3, sigma2, 1000)
```

```python
for mu,sigma,s,color in zip([mu1,mu2,mu3],[sigma1,sigma2,sigma3],[s1,s2,s3],['b','g','r']):
    count, bins, ignored = plt.hist(s, 30, density=True)
    plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
                       np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
                 linewidth=2, color=color)
```

```python
solar_uniform = np.random.uniform(low=250,high=1000,size=1000)
solar_exp = 300 + np.random.exponential(250,10000)
```

```python
_=plt.hist(solar_uniform, bins=bin, edgecolor='blue') 
```

```python
size = 1000
bin = np.arange(0,1000,1)

plt.hist(solar_exp, bins=bin, edgecolor='blue') 
plt.title("Exponential Distribution") 
plt.xlabel('wavelength (nm)')
plt.show()
```

```python
solar_poisson = 300 + np.random.lognormal(mean=np.log(500),sigma=2,size=10000)
plt.hist(solar_poisson, bins=bin, edgecolor='blue') 
plt.title("Lognormal Distribution") 
plt.xlabel('wavelength (nm)')
plt.show()
```
